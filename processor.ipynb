```python
import pandas as pd
import unicodedata
from random import shuffle

df = pd.read_csv("vicorpus.csv")
```

```python
def normalize(s):
    s = s.lower()
    normal = unicodedata.normalize('NFD', s)
    return ''.join([c for c in normal if unicodedata.category(c) != 'Mn'])
```

```python
tokens = (
    df['Content']
    .astype(str)
    .apply(lambda x: [
        normalize(t) for t in x.split(" ")
        if "|" not in x and len(t) == 5 and all(c in "abcdefghijklmnopqrstuvwxyzø" for c in normalize(t))
    ])
)
```

```python
all_tokens = [t for sublist in tokens for t in sublist]
token_counts = pd.Series(all_tokens).value_counts()
token_counts.sort_values(ascending=False)
```

```python
word_list = ([s for s in token_counts[:200].index])
shuffle(word_list)
print("\",\n\"".join(word_list))
```
